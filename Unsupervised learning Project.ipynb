```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn as sk
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df =pd.read_csv('Global Terrorism Data.csv',encoding='latin-1')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
data = df[['latitude', 'longitude', 'region_txt', 'attacktype1_txt']]

# Remove missing values
data = data.dropna()

# Remove duplicates
data = data.drop_duplicates()

print("Shape:", data.shape)
```

```python
df.head()
```

```python
data = data.sample(n=20000, random_state=42)
```

```python
data_encoded = pd.get_dummies(data, columns=['region_txt', 'attacktype1_txt'])
```

```python
data_encoded 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalr= StandardScaler()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_encoded)
```

```python
scalr.fit(data_encoded)
```

```python
scaled_data
```

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
data['KMeans_Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
data['KMeans_Cluster']
```

```python
sns.scatterplot(x=data['longitude'], y=data['latitude'], hue=data['KMeans_Cluster'])
plt.title('K-Means Clusters')
plt.show()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

print(pca.explained_variance_ratio_)
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=10)

data['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_data)

print(np.unique(data['DBSCAN_Cluster']))
```

```python
print(data['DBSCAN_Cluster'].value_counts())
```

```python
sns.scatterplot(x=data['longitude'], y=data['latitude'], hue=data['DBSCAN_Cluster'])
plt.title('DBSCAN Hotspots')
plt.show()
```

# **Best Algorithm**

**DBSCAN** is the most suitable algorithm.
